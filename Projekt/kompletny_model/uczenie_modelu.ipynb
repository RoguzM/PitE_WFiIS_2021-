```python

import matplotlib
matplotlib.use("Agg")
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import SGD
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from imutils import build_montages
import matplotlib.pyplot as plt
import numpy as np
import argparse
import cv2
import MiniVggnet
from tensorflow.keras.models import load_model
import os
```

```python

data = []
labels = []

for dir in os.listdir("./by_class"):
    bytes_object = bytes.fromhex(dir)
    ascii_string = bytes_object.decode("ASCII")

    if(dir):
        
        for dir1 in os.listdir("./by_class/"+dir):
            if(os.path.isdir("./by_class/"+dir+"/"+dir1)):
                if(dir1.__len__()<=5):
                    print(dir1)
                    for file in os.listdir("./by_class/"+dir+"/"+dir1):
                        img = cv2.imread(os.path.join("./by_class/"+dir+"/"+dir1, file))
                        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
                        data.append(img)
                        labels.append(ord(ascii_string))
       
            

       
        
             
```

```python

data = [cv2.resize(image, (32, 32)) for image in data]
data = np.array(data, dtype="float32")
```

```python
data /= 255.0
```

```python
aug = ImageDataGenerator(
	rotation_range=10,
	zoom_range=0.05,
	width_shift_range=0.1,
	height_shift_range=0.1,
	shear_range=0.15,
	horizontal_flip=False,
	fill_mode="nearest")
```

```python
nazwy=np.unique(labels)
print(nazwy)
```

```python
nazwy= np.array(nazwy, dtype="int")
print(nazwy.__len__())
```

```python
opt = SGD(learning_rate=0.01)

model=MiniVggnet.MiniVGGNet.build(width=32,height=32,depth=1,classes=62)
model.compile(loss="categorical_crossentropy",optimizer=opt,metrics=["accuracy"])
```

```python
labels=np.array(labels, dtype="int")
```

```python

le = LabelBinarizer()
labels = le.fit_transform(labels)


counts = labels.sum(axis=0)


classTotals = labels.sum(axis=0)
classWeight = {}


for i in range(0, len(classTotals)):
	classWeight[i] = classTotals.max() / classTotals[i]
```

```python
data = np.expand_dims(data, axis=-1)
```

```python

(trainX, testX, trainY, testY) = train_test_split(data,labels, test_size=0.20, stratify=labels, random_state=42)
print(data.shape)
print(labels.shape)
```

```python
print("[INFO] training network...")
H = model.fit(aug.flow(data, labels, batch_size=32),validation_data=(data, labels), steps_per_epoch=len(data) // 32,epochs=50, verbose=1)
```

```python


print("[INFO] evaluating network...")
print(testX.__len__())
print(testY.__len__())
print(nazwy.__len__())
predictions = model.predict(testX, batch_size=128)
```

```python
nazwy=[chr(l)for l in nazwy]
print(nazwy)
```

```python
print(classification_report(testY.argmax(axis=1),predictions.argmax(axis=1), target_names=nazwy))
```

```python
model.save("kompletny.model", save_format="h5")
N = np.arange(0, 50)
plt.style.use("ggplot")
plt.figure()
plt.plot(N, H.history["loss"], label="train_loss")
plt.plot(N, H.history["val_loss"], label="val_loss")
plt.title("Training Loss and Accuracy")
plt.xlabel("Epoch #")
plt.ylabel("Loss/Accuracy")
plt.legend(loc="lower left")
plt.savefig("plot.png")
```

```python
images = []

for i in np.random.choice(np.arange(0, len(testY)), size=(49,)):

	probs = model.predict(testX[np.newaxis, i])
	prediction = probs.argmax(axis=1)
	label = nazwy[prediction[0]]
	
	image = (testX[i] * 255).astype("uint8")
	color = (0, 255, 0)
	
	if prediction[0] != np.argmax(testY[i]):
		color = (0, 0, 255)

	image = cv2.merge([image] * 3)
	image = cv2.resize(image, (96, 96), interpolation=cv2.INTER_LINEAR)
	cv2.putText(image, label, (5, 20), cv2.FONT_HERSHEY_SIMPLEX, 0.75,color, 2)
	images.append(image)

montage = build_montages(images, (96, 96), (7, 7))[0]

cv2.imshow("OCR Results", montage)
cv2.waitKey(0)
```
